### Train loss

```python
fig,ax = plt.subplots(figsize=(12,8))

#ax.set_title('Multi30K Training loss curve', fontsize=15)
df_e6d6.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='dotted')
df_e3d3.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='dashed')
df_e3d2.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='dashdot')
df_e3d1.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='dotted')
df_e8d1.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='solid')
df_cmp_e3d3.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='dashdot')
df_cmp_e8d1.plot.line(ax=ax,x='epoch',y=["train_loss"],linestyle='solid')
ax.set_ylabel('Value')
ax.set_xlabel('Epoch')
ax.legend(["6/6","3/3","3/2","3/1","8/1","3/3 Comp. Dec.","8/1 Comp. Dec."])
fig.savefig("/content/drive/MyDrive/results/multi30k_trainloss.png",format='png', dpi=400)
plt.show()
```
